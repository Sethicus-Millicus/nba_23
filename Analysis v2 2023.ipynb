```python
def get_shooting_type(df_slice):
    x={}
    

    x['MIN/G'] = df_slice['MIN'].sum()/df_slice['MIN'].count()
    x['WO_3P_(%)'] = df_slice['3PM_WO'].sum()/df_slice['3PA_WO'].sum()
    x['WO_3P_rate'] = (df_slice['3PA_WO'].sum()/df_slice['MIN'].sum())*36
    x['O_3P_(%)'] = df_slice['3PM_O'].sum()/df_slice['3PA_O'].sum()
    x['O_3P_rate'] = (df_slice['3PA_O'].sum()/df_slice['MIN'].sum())*36
    x['T_3P_(%)'] = df_slice['3PM_T'].sum()/df_slice['3PA_T'].sum()
    x['T_3P_rate'] = (df_slice['3PA_T'].sum()/df_slice['MIN'].sum())*36
    x['VT_3P_(%)'] = df_slice['3PM_VT'].sum()/df_slice['3PA_VT'].sum()
    x['VT_3P_rate'] = (df_slice['3PA_VT'].sum()/df_slice['MIN'].sum())*36
    x['Tot_3P_(%)'] = df_slice['3PM'].sum()/df_slice['3PA'].sum()
    x['Tot_3P_rate'] = (df_slice['3PA'].sum()/df_slice['MIN'].sum())*36
    x['CS/PU'] = df_slice['3PA_CS'].sum()/(df_slice['3PA_CS'].sum()+df_slice['3PA_PULLUP'].sum())
    x['HO/36min'] = df_slice['HO/Min'].max()*36
    x['OFFs/36min'] = df_slice['OFF/Min'].max()*36
    
    return pd.Series(x, index=['MIN/G','Tot_3P_(%)','Tot_3P_rate','CS/PU','HO/36min','OFFs/36min'])
```

```python
def get_shooting_prob(df_slice):
    x={}
    
    x['WO_paint_(%)'] = df_slice['2FGM_WO_paint'].sum()/df_slice['2FGA_WO_paint'].sum()
    x['WO_paint_rate'] = (df_slice['2FGA_WO_paint'].sum()/df_slice['MIN'].sum())*36
    x['O_paint_(%)'] = df_slice['2FGM_O_paint'].sum()/df_slice['2FGA_O_paint'].sum()
    x['O_paint_rate'] = (df_slice['2FGA_O_paint'].sum()/df_slice['MIN'].sum())*36
    x['T_paint_(%)'] = df_slice['2FGM_T_paint'].sum()/df_slice['2FGA_T_paint'].sum()
    x['T_paint_rate'] = (df_slice['2FGA_T_paint'].sum()/df_slice['MIN'].sum())*36
    x['VT_paint_(%)'] = df_slice['2FGM_VT_paint'].sum()/df_slice['2FGA_VT_paint'].sum()
    x['VT_paint_rate'] = (df_slice['2FGA_VT_paint'].sum()/df_slice['MIN'].sum())*36
    x['Tot_paint_(%)'] = df_slice['2FGM_paint'].sum()/df_slice['2FGA_paint'].sum()
    x['Tot_paint_rate'] = (df_slice['2FGA_paint'].sum()/df_slice['MIN'].sum())*36
    
    x['WO_mid_(%)'] = df_slice['2FGM_WO_10'].sum()/df_slice['2FGA_WO_10'].sum()
    x['WO_mid_rate'] = (df_slice['2FGA_WO_10'].sum()/df_slice['MIN'].sum())*36
    x['O_mid_(%)'] = df_slice['2FGM_O_10'].sum()/df_slice['2FGA_O_10'].sum()
    x['O_mid_rate'] = (df_slice['2FGA_O_10'].sum()/df_slice['MIN'].sum())*36
    x['T_mid_(%)'] = df_slice['2FGM_T_10'].sum()/df_slice['2FGA_T_10'].sum()
    x['T_mid_rate'] = (df_slice['2FGA_T_10'].sum()/df_slice['MIN'].sum())*36
    x['VT_mid_(%)'] = df_slice['2FGM_VT_10'].sum()/df_slice['2FGA_VT_10'].sum()
    x['VT_mid_rate'] = (df_slice['2FGA_VT_10'].sum()/df_slice['MIN'].sum())*36
    x['Tot_mid_(%)'] = df_slice['2FGM_10'].sum()/df_slice['2FGA_10'].sum()
    x['Tot_mid_rate'] = (df_slice['2FGA_10'].sum()/df_slice['MIN'].sum())*36
    
    x['WO_3P_(%)'] = df_slice['3PM_WO'].sum()/df_slice['3PA_WO'].sum()
    x['WO_3P_rate'] = (df_slice['3PA_WO'].sum()/df_slice['MIN'].sum())*36
    x['O_3P_(%)'] = df_slice['3PM_O'].sum()/df_slice['3PA_O'].sum()
    x['O_3P_rate'] = (df_slice['3PA_O'].sum()/df_slice['MIN'].sum())*36
    x['T_3P_(%)'] = df_slice['3PM_T'].sum()/df_slice['3PA_T'].sum()
    x['T_3P_rate'] = (df_slice['3PA_T'].sum()/df_slice['MIN'].sum())*36
    x['VT_3P_(%)'] = df_slice['3PM_VT'].sum()/df_slice['3PA_VT'].sum()
    x['VT_3P_rate'] = (df_slice['3PA_VT'].sum()/df_slice['MIN'].sum())*36
    x['Tot_3P_(%)'] = df_slice['3PM'].sum()/df_slice['3PA'].sum()
    x['Tot_3P_rate'] = (df_slice['3PA'].sum()/df_slice['MIN'].sum())*36
    #x['CS_3P_rate'] = (df_slice['3PA_CS'].sum()/df_slice['MIN'].sum())*36
    #x['PU_3P_rate'] = (df_slice['3PA_PULLUP'].sum()/df_slice['MIN'].sum())*36
    
    return pd.Series(x, index=['WO_paint_(%)','WO_paint_rate','O_paint_(%)','O_paint_rate',
                              'T_paint_(%)','T_paint_rate','VT_paint_(%)','VT_paint_rate','Tot_paint_(%)','Tot_paint_rate',
                              'WO_mid_(%)','WO_mid_rate','O_mid_(%)','O_mid_rate',
                              'T_mid_(%)','T_mid_rate','VT_mid_(%)','VT_mid_rate','Tot_mid_(%)','Tot_mid_rate',
                              'WO_3P_(%)','WO_3P_rate','O_3P_(%)','O_3P_rate',
                              'T_3P_(%)','T_3P_rate','VT_3P_(%)','VT_3P_rate','Tot_3P_(%)','Tot_3P_rate'])
```

```python
def get_shooting_ptypes(df_slice):
    x={}
    
    
    x['PTS/36Min'] = (df_slice['PTS'].sum()/df_slice['MIN'].sum())*36
    x['Paint_PTS'] = (df_slice['2FGM_paint'].sum()/df_slice['MIN'].sum())*36
    x['OPEN_MID_RATE'] = ((df_slice['2FGA_WO_10'].sum()+df_slice['2FGA_O_10'].sum())/df_slice['MIN'].sum())*36
    x['OPEN_3P_RATE'] = ((df_slice['3PA_WO'].sum()+df_slice['3PA_O'].sum())/df_slice['MIN'].sum())*36
    x['3PM/36Min'] = ((df_slice['3PM'].sum())/df_slice['MIN'].sum())*36
    x['PTS_pg'] = (df_slice['PTS'].mean())
    x['3PM_pg'] = (df_slice['3PM'].mean())
    #x['CS_3P_rate'] = (df_slice['3PA_CS'].sum()/df_slice['MIN'].sum())*36
    #x['PU_3P_rate'] = (df_slice['3PA_PULLUP'].sum()/df_slice['MIN'].sum())*36
    
    return pd.Series(x, index=['PTS/36Min','3PM/36Min','PTS_pg','3PM_pg','Paint_PTS','OPEN_MID_RATE','OPEN_3P_RATE'])
```

```python
def get_shooting_type(df_slice):
    x={}
    

    x['MIN/G'] = df_slice['MIN'].sum()/df_slice['MIN'].count()
    x['WO_3P_(%)'] = df_slice['3PM_WO'].sum()/df_slice['3PA_WO'].sum()
    x['WO_3P_rate'] = (df_slice['3PA_WO'].sum()/df_slice['MIN'].sum())*36
    x['O_3P_(%)'] = df_slice['3PM_O'].sum()/df_slice['3PA_O'].sum()
    x['O_3P_rate'] = (df_slice['3PA_O'].sum()/df_slice['MIN'].sum())*36
    x['T_3P_(%)'] = df_slice['3PM_T'].sum()/df_slice['3PA_T'].sum()
    x['T_3P_rate'] = (df_slice['3PA_T'].sum()/df_slice['MIN'].sum())*36
    x['VT_3P_(%)'] = df_slice['3PM_VT'].sum()/df_slice['3PA_VT'].sum()
    x['VT_3P_rate'] = (df_slice['3PA_VT'].sum()/df_slice['MIN'].sum())*36
    x['Tot_3P_(%)'] = df_slice['3PM'].sum()/df_slice['3PA'].sum()
    x['Tot_3P_rate'] = (df_slice['3PA'].sum()/df_slice['MIN'].sum())*36
    x['CS/PU'] = df_slice['3PA_CS'].sum()/(df_slice['3PA_CS'].sum()+df_slice['3PA_PULLUP'].sum())
    x['HO/36min'] = df_slice['HO/Min'].max()*36
    x['OFFs/36min'] = df_slice['OFF/Min'].max()*36
    
    return pd.Series(x, index=['MIN/G','Tot_3P_(%)','Tot_3P_rate','CS/PU','HO/36min','OFFs/36min'])
```

```python
def get_reb_stat(df_slice):
    x={}
    x['REB'] = df_slice['REB'].sum()
    x['UNCONTESTED_REB%'] = (df_slice['REB'].sum()-df_slice['Contested REB'].sum())/df_slice['REB'].sum()
    x['UNCONTESTED_REB/36MIN'] = ((df_slice['REB'].sum()-df_slice['Contested REB'].sum())/df_slice['MIN'].sum())*36
    x['REB/36MIN'] = (df_slice['REB'].sum()/df_slice['MIN'].sum())*36
    #x['DRIVE_SUCCESS_RATE'] = (df_slice['FGM'].sum()+(df_slice['FTA'].sum()/2)+df_slice['AST'].sum())/df_slice['DRIVES'].sum()
    
    return pd.Series(x, index=['REB','UNCONTESTED_REB%','REB/36MIN','UNCONTESTED_REB/36MIN'])

def get_reb_stats_player(df_slice):
    x={}
    x['REB'] = df_slice['REB'].sum()
    x['REB/G'] = df_slice['REB'].sum()/df_slice['GP'].sum()
    x['UNCONTESTED_REB/G'] = (df_slice['REB'].sum()-df_slice['Contested REB'].sum())/df_slice['GP'].sum()
    x['UNCONTESTED_REB/36MIN'] = ((df_slice['REB'].sum()-df_slice['Contested REB'].sum())/df_slice['MIN'].sum())*36
    x['UNCONTESTED_REB%'] = ((df_slice['REB'].sum()-df_slice['Contested REB'].sum())/df_slice['REB'].sum())
    x['REB/36MIN'] = (df_slice['REB'].sum()/df_slice['MIN'].sum())*36
    #x['DRIVE_SUCCESS_RATE'] = (df_slice['FGM'].sum()+(df_slice['FTA'].sum()/2)+df_slice['AST'].sum())/df_slice['DRIVES'].sum()
    
    return pd.Series(x, index=['REB','REB/G','UNCONTESTED_REB/G','UNCONTESTED_REB%','UNCONTESTED_REB/36MIN','REB/36MIN'])

def get_reb_stat_team(df_slice):
    x={}
    x['REB'] = df_slice['REB'].sum()
    x['UNCONTESTED_REB%'] = (df_slice['REB'].sum()-df_slice['Contested REB'].sum())/df_slice['REB'].sum()
    x['UNCONTESTED_REB/36MIN'] = ((df_slice['REB'].sum()-df_slice['Contested REB'].sum())/df_slice['MIN'].sum())*36
    x['REB/36MIN'] = (df_slice['REB'].sum()/df_slice['MIN'].sum())*36
    x['REB +/-'] = ((((df_slice['REB']-df_slice['Contested REB'])/df_slice['MIN'])*36)-df_slice['UNCONTESTED_REB/36MIN']).sum()
    
    
    return pd.Series(x, index=['REB','UNCONTESTED_REB%','UNCONTESTED_REB/36MIN','REB/36MIN','REB +/-'])

def get_tm_uncontested_reb(df_slice):
    x={}
    
    
    x['TEAMS_UC_REB'] = (df_slice['REB'].sum()-df_slice['Contested REB'].sum())

    
    return pd.Series(x, index=['TEAMS_UC_REB'])

def get_uncontested_share(df_slice):
    x={}
    
    
    x['UC_REB_SHARE'] = df_slice['Contested REB'].sum()/df_slice['TEAMS_UC_REB'].sum()

    
    return pd.Series(x, index=['UC_REB_SHARE'])
```

```python
def get_rebs_(df_slice):
    x={}
    x['REB_pg'] = (df_slice['REB']).mean()
    x['U_reb_pg'] = (df_slice['REB'].sum()-df_slice['Contested REB'].sum())/(df_slice['GAME_DATE']).count()
    #x['PTS_Over'] = (((df_slice['PTS']/df_slice['MIN'])*36)>df_slice['PTS/36Min']).count()
    
    
    return pd.Series(x, index=['REB_pg','U_reb_pg'])
```

```python
def get_reb_ptypes_dif(df_slice):
    x={}
    
    x['Games'] = (df_slice['GAME_DATE']).count()
    #x['PTS_Over'] = (((df_slice['PTS']/df_slice['MIN'])*36)>df_slice['PTS/36Min']).count()
    x['rebs_dif'] = ((df_slice['REB'])-df_slice['REB_pg']).sum()
    x['rebs_OVER'] = ((df_slice['REB'])>(df_slice['REB_pg']+2)).sum()/(df_slice['GAME_DATE']).count()
    x['UNC_rebs_dif'] = (((df_slice['REB'])-(df_slice['Contested REB']))-df_slice['U_reb_pg']).sum()
    x['UNC_rebs_OVER'] = (((df_slice['REB'])-(df_slice['Contested REB']))>(df_slice['U_reb_pg']+2)).sum()/(df_slice['GAME_DATE']).count()
    return pd.Series(x, index=['Games','rebs_dif','rebs_OVER','UNC_rebs_dif','UNC_rebs_OVER'])
```
