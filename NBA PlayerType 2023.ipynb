```python

def get_shooting_shares(df_slice):
    x={}
    x['Paint_Share'] = df_slice['Paint_FGA'].sum()/(df_slice['Paint_FGA'].sum()+df_slice['Mid_FGA'].sum()+df_slice['3PA'].sum())
    x['Paint_Share'] = df_slice['Paint_FGA'].sum()/(df_slice['Paint_FGA'].sum()+df_slice['Mid_FGA'].sum()+df_slice['3PA'].sum())
    x['Mid_Share'] = df_slice['Mid_FGA'].sum()/(df_slice['Paint_FGA'].sum()+df_slice['Mid_FGA'].sum()+df_slice['3PA'].sum())
    x['3PA_Share'] = df_slice['3PA'].sum()/(df_slice['Paint_FGA'].sum()+df_slice['Mid_FGA'].sum()+df_slice['3PA'].sum())
    x['Paint/36MIN'] = (df_slice['Paint_FGA'].sum()/df_slice['MIN'].sum())*36
    x['Mid/36MIN'] = (df_slice['Mid_FGA'].sum()/df_slice['MIN'].sum())*36
    x['3PA/36MIN'] = (df_slice['3PA'].sum()/df_slice['MIN'].sum())*36
    x['Drives/36MIN'] = (df_slice['FGA_DRIVES'].sum()/df_slice['MIN'].sum())*36
    
    return pd.Series(x, index=['Paint_Share','Mid_Share','3PA_Share','Paint/36MIN','Mid/36MIN','3PA/36MIN','Drives/36MIN'])
```

```python
shooting_shares = shooting_.groupby(['PLAYER']).apply(get_shooting_shares).reset_index()
```
